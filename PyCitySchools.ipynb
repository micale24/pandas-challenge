```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data = "./pandas-challenge/PyCitySchools/Resources/schools_complete.csv"
student_data = "./pandas-challenge/PyCitySchools/Resources/students_complete.csv"

# # Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data)
student_data = pd.read_csv(student_data)

# # Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
#Calculate the total number of students
total_students = school_data_complete['student_name'].count()

#Calculate the total number of schools
sum_total_schools  = len(school_data_complete['school_name'].unique())

#Calculate the total budget
total_budget = school_data_complete['budget'].unique().sum()

#Calculate the average math score
avg_math_score = round(school_data_complete['math_score'].mean(),5)

#Calculate the average reading score
avg_reading_score = round(school_data_complete['reading_score'].mean(),5)

#Calculate the percentage of students with a passing math score (70 or greater)
math_pass = school_data_complete[['math_score']]

students_passing_math = math_pass.loc[math_pass['math_score']>=70]

num_students_pmath = students_passing_math.count()

#Calculate the percentage of students with a passing math score (70 or greater)
read_pass = school_data_complete[['reading_score']]

students_passing_read = read_pass.loc[read_pass['reading_score']>=70]

num_students_pread = students_passing_read.count()

percentage_pass_read = round((num_students_pread/total_students)*100,6)

#Calculate the percentage of students who passed math and reading (% Overall Passing)
pass_rm = school_data_complete[['math_score','reading_score']]

PASS = pass_rm[pass_rm.iloc[:,:2]>=70]

stu_passed_readmath = PASS.dropna()
smart_stu = len(stu_passed_readmath)

overall_passing = round((smart_stu/total_students)*100,6)
```

```python
#Direct Summary Data Frame
direct_summary = pd.DataFrame ({'Total Schools':[sum_total_schools],
                             'Total Students':[total_students],
                             'Total Budget':[total_budget],
                             'Average Math Score':[avg_math_score],
                             'Average Reading Score':[avg_reading_score],
                             '% Passing Math':[percentage_pass_math[0]],
                             '% Passing Reading':[percentage_pass_read[0]],
                             '% Overall Passing':[overall_passing]})

#Optional formating 
direct_summary['Total Budget']= direct_summary['Total Budget'].astype(float).map("${:,.2f}".format)
direct_summary['Total Students']= direct_summary['Total Students'].astype(float).map("{:,.0f}".format)
```
